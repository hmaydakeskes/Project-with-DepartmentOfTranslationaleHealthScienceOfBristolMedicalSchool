# Concaténer les DataFrames en un seul

```python
from functools import reduce

# Liste de toutes vos dataframes
dfs = [AC3A_subset, AC4A_subset, AC5A_subset, AC6A_subset, AC7A_subset]

# Utilisez la fonction reduce pour fusionner toutes les dataframes
final_df = reduce(lambda left,right: pd.merge(left,right,on='Probe Set Name'), dfs)
```
